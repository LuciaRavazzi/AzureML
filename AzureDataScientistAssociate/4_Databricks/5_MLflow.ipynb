### Track experiments with MLflow

MLflow holds four components:
- MLflow tracking tracks metrics and metadata of models. the execution of a code is known as _run_ which is defined by _parameters_, which in this case are the hyperparamters, _metrics_, generated by the model and _artifacts_, namely the output files. An _experiment_ is a set of multiple runs.
- MLflow Projects allows us to pack our code and dependencies in order to be reusable by other data scientisti. It consists of the code and the setting of environment.
- MLflow Models offers a standardized way of ackaging models for disitribution.
- MLflow Model Registry allows data scientist to register models in a registry. They can be versioned. They can be belong to different stages like Staging, Production or Archive.
